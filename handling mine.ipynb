```python
import pandas as pd
#Reading the csv file
df = pd.read_csv('Mine.csv')
#Reading empty cells
print(df.isnull().sum())
```

```python
#Filling the empty cells with the mean of the column
for col in ['Pulse', 'Maxpulse', 'Calories' ]:
	df[col].fillna(df[col].mean(), inplace=True)
print(df.to_string())
 
```

```python
from datetime import datetime

df['Date'] = pd.to_datetime(df['Date'], format='mixed')
print(df.to_string())
```

```python
#Duplicate rows
df.drop_duplicates(inplace=True)
print(df.to_string())
```

```python
#Wrong data 
for x in df.index:
    if df.loc[x, "Duration"] > 60:
        df.loc[x, "Duration"] = 60
print(df.to_string())
```
